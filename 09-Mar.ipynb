### Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

The Probability Mass Function (PMF) and Probability Density Function (PDF) are two important concepts in probability theory and statistics.

The PMF is a function that describes the probability of a discrete random variable taking on a specific value. It gives the probability of each value that the random variable can take. The sum of probabilities for all possible values of the random variable equals 1. For example, consider the tossing of a fair coin. The PMF for this experiment will give a probability of 0.5 for both the outcomes "heads" and "tails".

The PDF is a function that describes the probability of a continuous random variable taking on a specific value or range of values. It represents the relative likelihood of observing a particular outcome within a range of possible outcomes, as opposed to a specific outcome. The area under the curve of the PDF over a particular interval gives the probability that the random variable falls within that interval. For example, consider the distribution of heights of adults in a population. The PDF for this distribution gives the probability density of observing a height within a certain range, such as 5'8" to 5'9".

To illustrate the difference between the PMF and PDF, consider the example of rolling a fair six-sided die. The PMF for this experiment will give a probability of 1/6 for each of the six possible outcomes (rolling a 1, 2, 3, 4, 5, or 6). The PDF for this experiment would be a uniform distribution, which is a horizontal line segment over the interval [1, 6], with a height of 1/6 (since each outcome has an equal probability of occurring). The area under the curve of the PDF between any two points in the interval [1, 6] gives the probability of rolling a number within that range.

In summary, the PMF is used to describe the probability of discrete random variables, while the PDF is used to describe the probability of continuous random variables.

The shape of the distribution is determined by these two parameters.

The area under the curve represents the probability of a random variable taking a certain value or falling within a certain range of values.

When μ ≠ 0 and/or σ ≠ 1, the distribution is called a normal distribution with mean μ and standard deviation σ. The shape of the distribution is similar to the standard normal distribution, but it may be shifted to the left or right, and/or stretched or compressed. The mean determines the location of the center of the distribution, while the standard deviation determines the spread or variability of the distribution.

Conversely, if the mean is less than 0, the distribution will be shifted to the left, and if the standard deviation is small, the distribution will be more concentrated around the mean. The shape of the normal distribution is important because it allows us to make probabilistic statements about the likelihood of a particular outcome. The normal distribution is also useful in statistical inference, where we use sample data to make inferences about the population mean and variance.

### Q7: Explain uniform Distribution with an example.

The uniform distribution is a continuous probability distribution that describes a situation where every value in a given range is equally likely to occur. In other words, the probability density function of a uniform distribution is constant within a certain range and zero outside of that range.

For example, let's consider a fair six-sided die. The possible outcomes when rolling the die are the integers 1, 2, 3, 4, 5, and 6. If we assume that each of these outcomes is equally likely, then we can model this situation using a uniform distribution over the interval [1, 6]. The probability density function of the uniform distribution is given by:

f(x) = 1/6, for x in [1, 6]
f(x) = 0, otherwise

This means that every value in the range [1, 6] has an equal probability of occurring, and the probability of any value outside of this range is zero. For example, the probability of rolling a 3 is 1/6, the probability of rolling a 7 is 0, and the probability of rolling a number between 2 and 4 (inclusive) is (4-2+1)/6 = 1/3.

The uniform distribution is useful in situations where we have no prior knowledge or assumptions about the likelihood of different values within a given range. It is also commonly used in simulations and random number generation, where we want to generate values that are equally likely to occur within a certain range.






### Q8: What is the z score? State the importance of the z score.

The z-score, also known as the standard score, is a measure of how many standard deviations a data point is from the mean of a distribution. It is calculated as:

z = (x - μ) / σ

where x is the data point, μ is the mean of the distribution, and σ is the standard deviation of the distribution.

The z-score is important because it allows us to compare values from different distributions that may have different units or scales. By standardizing the values to a common scale, we can make meaningful comparisons and draw conclusions about the relative position of a value within its distribution.

For example, suppose we have two distributions of test scores: one for a math test with a mean of 70 and a standard deviation of 10, and another for an English test with a mean of 80 and a standard deviation of 5. If we want to compare a score of 75 in math to a score of 85 in English, we can use the z-score to standardize the values:

z_math = (75 - 70) / 10 = 0.5

z_english = (85 - 80) / 5 = 1

This tells us that a score of 75 in math is half a standard deviation above the mean, while a score of 85 in English is one standard deviation above the mean. Thus, we can conclude that the score of 85 in English is relatively higher than the score of 75 in math, even though the raw scores have different units and scales.

In addition to comparing values from different distributions, the z-score is also used in hypothesis testing and confidence interval estimation. It allows us to calculate the probability of observing a value or a set of values in a distribution, assuming that the distribution is normal or approximately normal.

### Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

The Central Limit Theorem (CLT) is a fundamental concept in statistics that describes the behavior of the means of a large number of independent and identically distributed random variables, regardless of the distribution of the underlying variables.
The Central Limit Theorem states that if we take a large number of random samples from any distribution with a finite mean and variance, the sampling distribution of the means will be approximately normally distributed, regardless of the shape of the original distribution. This means that the means of the samples will tend to cluster around the true population mean, and the spread of the distribution of means will be reduced as the sample size increases.
The significance of the Central Limit Theorem lies in its wide applicability in statistics and its practical implications. Some of the important applications of the Central Limit Theorem are:

1. Hypothesis testing - The Central Limit Theorem is used to test hypotheses about the means of populations, where the distribution of the population is unknown or non-normal.

2. Confidence intervals - The Central Limit Theorem is used to construct confidence intervals for the population mean, which helps in estimating the true mean of a population based on a sample mean.

3. Sample size determination - The Central Limit Theorem is used to determine the sample size required for a given level of precision in estimating the population mean.

4. Quality control - The Central Limit Theorem is used to monitor the quality of production processes by measuring the means of samples and comparing them to the population mean.

In summary, the Central Limit Theorem is an essential concept in statistics that enables us to make inferences about populations based on samples, regardless of the underlying distribution of the data. It has important applications in various fields, making it a crucial tool for data analysis and decision-making.

### Q10: State the assumptions of the Central Limit Theorem.

The Central Limit Theorem (CLT) is a fundamental theorem in statistics that describes the behavior of sample means and sample sums from a wide variety of probability distributions. The assumptions of the CLT are:

1. Independence: The observations in the sample must be independent of each other.
2. Sample Size: The sample size should be large enough, usually at least 30 observations. If the population distribution is highly skewed, a larger sample size may be required.
3. Finite Variance: The population from which the sample is drawn must have a finite variance. If the variance is infinite, the CLT does not apply.

These assumptions are critical to ensure that the sample mean or sum is approximately normally distributed. The CLT states that regardless of the shape of the population distribution, the sample mean or sum will be approximately normally distributed if the sample size is sufficiently large. The larger the sample size, the more closely the distribution of the sample mean or sum will resemble a normal distribution.

The CLT is widely used in statistical inference and hypothesis testing, as it provides a powerful tool for making inferences about population parameters from sample data. It allows us to make assumptions about the distribution of the sample mean or sum, which can then be used to construct confidence intervals, perform hypothesis tests, and make predictions about the behavior of the population.
