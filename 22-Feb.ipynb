## Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

In Pandas, a DataFrame is a two-dimensional labeled data structure that can hold data of different types (numeric, string, boolean, etc.) in columns. It is similar to a spreadsheet or SQL table, where data is organized in rows and columns. On the other hand, a Series is a one-dimensional labeled data structure that can hold data of different types in a single column.

```python
import pandas as pd

# Creating a Pandas Series
my_series = pd.Series([1, 2, 3, 4, 5])

# Creating a Pandas DataFrame
my_dataframe = pd.DataFrame({
    'Name': ['Pradosh', 'Kumar', 'Ankush', 'Rama', 'Dhruti'],
    'Age': [20, 25, 35, 40, 45],
    'Salary': [50000, 60000, 70000, 80000, 90000]
})

# Printing the Series and DataFrame
print("Pandas Series:\n", my_series)
print("\nPandas DataFrame:\n", my_dataframe)
```

## Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

Pandas provides a wide range of functions that can be used to manipulate data in a DataFrame. Here are some common functions that are frequently used:

1. head() and tail(): These functions are used to view the first or last n rows of a DataFrame. They are useful for quickly inspecting the data and checking if it has been read in correctly.

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
df.head()  # displays first 5 rows of the DataFrame
```

```python
df.tail() # displays last 5 rows of the DataFrame
```

2. describe(): This function generates descriptive statistics for the DataFrame, such as the count, mean, standard deviation, minimum, and maximum values of each column.

```python
df.describe()  # displays summary statistics for the DataFrame
```

3. groupby(): This function is used to group the data in the DataFrame based on one or more columns and perform aggregate calculations on the groups. It is useful for analyzing trends and patterns in the data.

```python
grouped_df = df.groupby(['Age', 'Sex']).sum()  # groups data by Age and sex
grouped_df.head()
```

4. fillna(): This function is used to fill missing values in the DataFrame with a specified value or method. It is useful for dealing with missing data, which is common in real-world datasets.

```python
df.fillna('Pradosh', inplace=True)# fills all missing values with 'Pradosh'
df
```

5. apply(): This function is used to apply a function to each element or row of the DataFrame. It is useful for performing complex calculations or transformations on the data.

```python
df['P'] = df['Pclass'].apply(lambda x: x*x)
df['P']
```

## Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

In Pandas, both Series and DataFrame are mutable, while Panel is immutable.

Series and DataFrame can be modified in place by changing or adding values to them. For example, you can change the value of a specific element in a Series or DataFrame using indexing or selection methods, such as loc[] or iloc[], and then assigning a new value to it.

Panel, on the other hand, is a three-dimensional data structure that is similar to a DataFrame but with an added panel axis. However, it has been deprecated since version 0.25.0 and is no longer recommended for use.

## Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
import pandas as pd

# create three Series
s1 = pd.Series([1, 2, 3, 4, 5], name='A')
s2 = pd.Series(['apple', 'banana', 'cherry', 'durian', 'elderberry'], name='B')
s3 = pd.Series([10.5, 20.0, 5.5, 30.0, 15.5], name='C')

# create a dictionary of Series
data = {'Fruit_list': s1, 'Fruit_Name': s2, 'Price': s3}

# create a DataFrame using the dictionary
df = pd.DataFrame(data)

# print the DataFrame
print(df)
```
