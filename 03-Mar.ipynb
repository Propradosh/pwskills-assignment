### Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a Python data visualization library based on Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Here are five types of plots that can be created using Seaborn, along with their uses:

1. Scatter plot: A scatter plot is a two-dimensional plot that shows the relationship between two variables. Seaborn provides a scatterplot() function that can be used to create scatter plots. Scatter plots are useful for identifying patterns and relationships in data.

2. Histogram: A histogram is a graphical representation of the distribution of a dataset. Seaborn provides a histplot() function that can be used to create histograms. Histograms are useful for understanding the distribution of data and identifying outliers.

3. Box plot: A box plot is a graphical representation of the distribution of a dataset through their quartiles. Seaborn provides a boxplot() function that can be used to create box plots. Box plots are useful for identifying outliers and extreme values.

4. Line plot: A line plot is a graphical representation of the relationship between two variables, where the variables are plotted on the x and y axes. Seaborn provides a lineplot() function that can be used to create line plots. Line plots are useful for visualizing trends over time or across categories.

5. Heatmap: A heatmap is a graphical representation of a matrix where the values are represented as colors. Seaborn provides a heatmap() function that can be used to create heatmaps. Heatmaps are useful for visualizing patterns and relationships in large datasets.






### Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

              Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create the line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
plt.title('Signal over Time for Different Events and Regions')
plt.show()
```

### Q3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
                     Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns

# Load the titanic dataset
sns.load_dataset("titanic")
```

```python
titanic = sns.load_dataset("titanic")
```

```python
# Create the first box plot
sns.boxplot(x="pclass", y="age", data=titanic)
plt.title('Showing the boxplot for pclass vs age')
plt.show()
```

```python
# Create the second box plot
sns.boxplot(x="pclass", y="fare", data=titanic)
plt.title('Showing the boxplot for pclass vs fare')
plt.show()
```

### Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns

# Load the diamonds dataset
sns.load_dataset('diamonds')
```

```python
diamonds = sns.load_dataset('diamonds')
```

```python
# Create a histogram of diamond prices with hue based on diamond cut
sns.histplot(data=diamonds, x='price', hue='cut')
plt.title('Distribution of Price for various diamond cut')
plt.show()
```

### Q6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
flights=sns.load_dataset('flights')
flights
```

```python
flights_matrix = flights.pivot(index="year", columns="month", values="passengers")
flights_matrix
```

```python
# Create a heatmap of the flights data
sns.heatmap(flights_matrix, 
            cmap="YlGnBu", 
            annot=True, 
            fmt="d", 
            linewidths=1,
            linecolor="red")
plt.title("Passengers per Month")
plt.show()
```
