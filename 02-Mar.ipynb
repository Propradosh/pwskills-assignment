### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a popular data visualization library for Python. It is used to create a variety of 2D plots and graphs, including line plots, scatter plots, bar plots, histograms, and more. Matplotlib is a powerful tool for visualizing data in Python and is widely used in scientific research, data analysis, and machine learning.

Matplotlib's Pyplot module provides a simple and easy-to-use interface for creating and customizing plots. Here are five types of plots that can be plotted using the Pyplot module of Matplotlib:

1. Line plot - A simple plot where each data point is connected with a line.
2. Scatter plot - A plot where individual data points are represented as points or markers.
3. Bar plot - A plot where data is represented as bars, with one axis showing the category and the other axis showing the value.
4. Histogram - A plot that displays the distribution of a continuous variable by dividing the data into a set of intervals (or "bins") and showing the frequency of data points in each bin.
5. Pie chart - A plot where data is represented as slices of a circle, with each slice showing the proportion of the data in a particular category.

Overall, Matplotlib is a powerful and versatile library for creating a wide range of visualizations, and its Pyplot module provides a convenient and easy-to-use interface for many common use cases.






### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

                                import numpy as np
                                np.random.seed(3)
                                x = 3 + np.random.normal(0, 2, 50)
                                y = 3 + np.random.normal(0, 2, len(x))
                                Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of plot that displays the relationship between two variables by showing their values as a collection of points on a 2D coordinate system. Each point in the scatter plot represents a single observation, and the position of the point reflects the values of the two variables being plotted.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot of X and Y')
plt.xlabel('X-axis values')
plt.ylabel('Y-axis values')
plt.show()
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
            Use the following data:
            import numpy as np
            For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
            For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
            For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
            For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])


The subplot() function in Matplotlib is used to create multiple plots within a single figure. It allows us to create a grid of plots with a specified number of rows and columns, and to specify the location of each plot within the grid using its row and column indices.

```python
import numpy as np

# Line 1
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
# Importing Matplotlib
import matplotlib.pyplot as plt

# Create four line plots in one figure using subplot()
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

# Plot 1 : Line 1
axes[0, 0].plot(x1, y1)
axes[0, 0].set_title('Line-1')

# Plot 2 : Line 2
axes[0, 1].plot(x2, y2)
axes[0, 1].set_title('Line-2')

# Plot 3 : Line 3
axes[1, 0].plot(x3, y3)
axes[1, 0].set_title('Line-3')

# Plot 4 : Line 4
axes[1, 1].plot(x4, y4)
axes[1, 1].set_title('Line-4')

# Add a common x-axis label and y-axis label
fig.text(0.5, 0.04, 'X-axis', ha='center')
fig.text(0.04, 0.5, 'Y-axis', va='center', rotation='vertical')

# Display the plot
plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
            import numpy as np
            company = np.array(["Apple", "Microsoft", "Google", "AMD"])
            profit = np.array([3000, 8000, 1000, 10000])

A bar plot, also known as a bar chart, is a type of plot that represents categorical data with rectangular bars, with the height or length of each bar proportional to the value of the category it represents.

Here are some specific reasons why bar plots are used:

1. Comparing quantities : Bar plots are used to compare the quantities of different categories, such as sales figures, revenue, or counts of items. By visually comparing the heights or lengths of the bars, it's easy to see which category has the largest or smallest value.

2. Tracking changes over time : Bar plots can also be used to track changes over time within a single category. For example, you might create a bar plot showing the revenue of a single department for each month of the year. By comparing the heights or lengths of the bars for each month, you can see how the revenue changes over time.

3. Showing relationships : Bar plots can also be used to show relationships between categories. For example, you might create a stacked bar plot showing the sales figures for different products in different regions. By stacking the bars on top of each other, you can see how the sales figures for each product compare across different regions.

4. Communicating results : Bar plots are a simple and effective way to communicate results to a non-technical audience. By presenting data in a visually appealing format, bar plots can help to convey the main points of your analysis quickly and clearly.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.bar(company, profit)
plt.title('Company Profit')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.show()
```

```python
# Horizontal bar plot
plt.barh(company, profit)
plt.title('Company Profit')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.show()
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
                        box1 = np.random.normal(100, 10, 200)
                        box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box and whisker plot, is a type of plot used to represent the distribution of a dataset. It shows the five-number summary of a dataset, which includes the minimum and maximum values, the lower and upper quartiles, and the median.

A box plot is typically used to identify and visualize the range and distribution of a dataset, and to identify outliers and extreme values.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Box plot
plt.boxplot([box1, box2])
plt.title('Box Plot')
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.ylabel('Value')
plt.show()
```
